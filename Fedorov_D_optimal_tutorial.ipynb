```python
from Fedorov_D_optimal import *
import numpy as np
```

```python
# Example Usage

# Imagine a chemical engineer tasked with optimizing the temperature for a chemical reaction 
# to maximize the yield. 
# We first suspect the relationship between temperature and yield is linear.

x_range = np.linspace(50, 150, 100)  # Temperature range from 50°C to 150°C
optimal_design, optimal_weights = d_optimal_design(x_range, degree=1, iterations=5, tolerance=0.1)

# Print the results
print("Optimal Design Points:\n", optimal_design)
print("Weights for Each Design Point:\n", optimal_weights)
```

```python
# This shows that the relationship is probably not linear
# We move on to try quadratic and cubic cases
```
